```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as tck
```

```python
df_excel = pd.read_excel("Pandas.xlsx", sheet_name="data")
df_excel['total_produksi_tahunan'] = df_excel.apply(lambda baris:baris['jumlah_produksi_sampah']*365, axis=1)
print("Data dari excel (Belum dikelompokkan)")
df_excel
```

```python
df_tahunan = df_excel.groupby(['nama_kabupaten_kota', 'tahun'])['total_produksi_tahunan'].sum().unstack()
print("Data dari df_tahunan (Total produksi di kategorikan berdasarkan Kabupaten/Kota dan Tahun Produksi)")
df_tahunan
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import pandas as pd
plt.figure(figsize=(9, 5))
plt.bar(df_excel['tahun'], df_excel['jumlah_produksi_sampah'], color='skyblue')
plt.xlabel('Jumlah Produksi Sampah (TON)')
plt.title('Jumlah Produksi Sampah per Tahun dan Kota')
plt.gca().yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:.0f}'))
plt.tight_layout()
plt.show()
```

```python
#soal No 2 
df_tahunan.plot(kind='bar', figsize=(18, 6), color=['purple', 'red', 'green', 'blue', 'yellow', 'black', 'pink'])
plt.ylabel('Jumlah Produksi (Ton)')
plt.xlabel('Nama Kota/Kabupaten')
plt.title('Jumlah Produksi Tahunan')
plt.gca().yaxis.set_major_formatter(tck.StrMethodFormatter('{x:,.0f}'))
plt.tight_layout()
plt.show()
```
